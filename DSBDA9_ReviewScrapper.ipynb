```python
# Create a review scrapper for any e-commerce website to fetch real time comments, 
# reviews, ratings, comment tags, customer name using Python.

# web scrapping-to pull out necessary/required data from website, 
# bcoz website contain large amount of data & actually it is not required.
```

```python
# Request access
# It is used to send http request to specified url to retrieve data using get method
import requests
```

```python
# Used for pulling data from webite in xml,html files
import bs4
```

```python
request1=requests.get('https://www.amazon.in/product-reviews/8172345666/ref=acr_dp_hist_5?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews#reviews-filter-bar')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
# Fetching reviews/comments
reviews=soup.find_all('span',{'class' : 'a-size-base review-text review-text-content'})
for review in reviews:
    print(review.get_text(),end="\n")
```

```python
# Fetching overall average rating
overall_ratings=soup.find('span',{'class' : 'a-size-medium a-color-base'})
print(overall_ratings.get_text())
```

```python
# Fetch customer name
customer_names=soup.find_all('span',{'class' : 'a-profile-name'})
for customer in customer_names:
    print(customer.get_text(),end="\n")
```

```python
tags=soup.find_all('a',{'class' : 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'})
for tag in tags:
    print(tag.get_text(),end="\n")
```
